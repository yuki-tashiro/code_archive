```python
from sklearn.preprocessing import MinMaxScaler

# スケーラーを初期化
scaler = MinMaxScaler()

# 'age'列をスケーリング
df_train['age'] = scaler.fit_transform(df_train[['age']])
df_train['TG'] = scaler.fit_transform(df_train[['TG']])
df_train['SBP'] = scaler.fit_transform(df_train[['SBP']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# プロットのサイズを設定
plt.figure(figsize=(10, 6))

# スケーリング後の 'age_scaled' 列のヒストグラムをプロット
sns.histplot(df_train['age'], color='red', label='Scaled Age', kde=True, bins=100)
sns.histplot(df_train['TG'], color='blue', label='Scaled TG', kde=True, bins=100)
sns.histplot(df_train['SBP'], color='green', label='Scaled SBP', kde=True, bins=100)

# ラベルとタイトル
plt.title('Distribution of Age (Original vs Scaled)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()

# グラフを表示
plt.show()
```
