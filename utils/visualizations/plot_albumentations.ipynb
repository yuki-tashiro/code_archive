# Albumentationsの概要

Albumentationsは、画像データの拡張（augmentation）を行うためのPythonライブラリです。機械学習やディープラーニングのモデルを訓練する際に、データの多様性を増やし、モデルの汎化性能を向上させるために使用されます。Albumentationsは、高速で柔軟性があり、さまざまな画像変換を簡単に適用できる点が特徴です。実際にKaggleなどのコンペティションで多くの人に利用されています。

# 1. 関数

```python
import cv2
import matplotlib.pyplot as plt

import albumentations as A

def show_image_albumentations(image_path, transform):
    """
    画像に任意のデータ拡張を適用して結果を確認する関数

    Args:
        image_path (str): 画像ファイルのパス
        transform (albumentations.Compose): Albumentationsのデータ拡張パイプライン

    Returns:
        None

    Example:
        >>> transform = A.Compose([A.HorizontalFlip(p=0.5)])
        >>> show_image_transform('path/to/image.jpg', transform)
    """
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    input = img.copy()
    print(f"Original Image Shape: {img.shape}")

    # 変換を適用
    transformed_image = transform(image=input)['image']

    # 元の画像と変換後の画像を並べて表示
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    # オリジナル画像の表示
    axes[0].imshow(img)
    axes[0].set_title(f"Original Image {img.shape[0]}x{img.shape[1]}x{img.shape[2]}")
    axes[0].axis('off')

    # 変換後の画像の表示
    axes[1].imshow(transformed_image)
    axes[1].set_title("Transformed Image: " + ', '.join([t.__class__.__name__ for t in transform.transforms]))
    axes[1].axis('off')

    # 両方の画像を表示
    plt.tight_layout()
    plt.show()
```

# 3. カスタムデータ拡張の例

```python
import numpy as np

# Canny法でエッジ検出するカスタム変換
class CannyEdgeDetection:
    def __init__(self, low_threshold, high_threshold):
        self.low_threshold = low_threshold
        self.high_threshold = high_threshold
        self.available_keys = {'image'}

    def __call__(self, image, **kwargs):
        image = image.astype(np.uint8)  # データ型を8ビットに変換
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # グレースケールに変換
        edges = cv2.Canny(gray_image, self.low_threshold, self.high_threshold, L2gradient=True)
        return {'image': edges}


# Sobelフィルタを適用するカスタム変換
class SobelEdgeDetection:
    def __init__(self, ksize=3, alpha=0.5, beta=0.5):
        self.ksize = ksize  # Sobelフィルタのカーネルサイズ
        self.alpha = alpha  # 横方向の重み
        self.beta = beta    # 縦方向の重み
        self.available_keys = {'image'}

    def __call__(self, image, **kwargs):
        image = image.astype(np.uint8)  # データ型を8ビットに変換

        # Sobelフィルタを適用
        img_sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=self.ksize)  # 横方向
        img_sobelx = cv2.convertScaleAbs(img_sobelx)

        img_sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=self.ksize)  # 縦方向
        img_sobely = cv2.convertScaleAbs(img_sobely)

        # 勾配の大きさを計算
        img_gradient = cv2.addWeighted(img_sobelx, self.alpha, img_sobely, self.beta, 0)

        return {'image': img_gradient}
```

```python
# データ拡張のパイプラインを定義
transform = A.Compose([
    A.CenterCrop(width=1200, height=1200),
    CannyEdgeDetection(low_threshold=25, high_threshold=60)  # Cannyエッジ検出を追加
])
show_image_albumentations(img_path, A.Compose([transform]))

transform = A.Compose([
    A.CenterCrop(width=1200, height=1200),
    SobelEdgeDetection(ksize=5, alpha=0.3, beta=0.3)  # ハイパーパラメータを指定
])
show_image_albumentations(img_path, A.Compose([transform]))
```
