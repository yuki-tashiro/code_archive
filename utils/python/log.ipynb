
reference
- logger: ログ出力の命令
- hander: 実施の出力の実行部分

ログのレベル
1. DEBUG: デバッグ情報, debug()
2. INFO: 一般情報, info()
3. WARNING: 警告, warning()
4. ERROR: エラー, 問題が起きた時のログ, error()
5. CRITICAL: クリティカル, システムが停止するような重大な問題, critical()

```python
from logging import getLogger, StreamHandler, Formatter, DEBUG

logger = getLogger(__name__)
handler = StreamHandler() # pythonの標準出力
handler.setLevel(DEBUG) # ログのレベルの設定
logger.setLevel(DEBUG) # ログのレベルの設定
logger.addHandler(handler) # ロガーにハンドラを追加
logger.debug("これはデバッグログです") # ログの出力
```

```python
formatter = Formatter(
    '[%(levelname)s] %(asctime)s %(message)s(%(filenames)s'
)
logger = getLogger(__name__)
handler = StreamHandler() # pythonの標準出力
handler.setLevel(DEBUG) # ログのレベルの設定
handler.setFormatter(formatter) # ログのフォーマットの設定
logger.setLevel(DEBUG) # ログのレベルの設定
logger.addHandler(handler) # ロガーにハンドラを追加
logger.debug("これはデバッグログです") # ログの出力
logger.error("これはエラーログです") # ログの出力
```

```python
from logging import getLogger, StreamHandler, Formatter, FileHandler, DEBUG

formatter = Formatter('[%(levelname)s] %(asctime)s %(message)s(%(filenames)s')
logger = getLogger(__name__)
# handler = StreamHandler() # pythonの標準出力
handler = FileHandler("test.log") # ファイル出力
handler.setLevel(DEBUG) # ログのレベルの設定
handler.setFormatter(formatter) # ログのフォーマットの設定
logger.setLevel(DEBUG) # ログのレベルの設定
logger.addHandler(handler) # ロガーにハンドラを追加
logger.debug("これはデバッグログです") # ログの出力
logger.error("これはエラーログです") # ログの出力
```
